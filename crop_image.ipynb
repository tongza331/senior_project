```python
import cv2
import os 
import glob
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
## Crop images
def crop_image(input):
    img = cv2.imread(input)
    cutted_img = img[55:1340, 340:1650]
    return cutted_img
```

```python
new = crop_image('./2021_11_29__15_20_00_HDR.jpg')
cv2.imwrite('./test1.jpg',new)
```

```python
path_data_2 = './class_cloud/8/'
sum_img = 0
for file in os.listdir(path_data_2):
    files = glob.glob(path_data_2+file)
    for file in files:
        filename = file.split('/')
        filename = file[16:36] + '_crop'
        new_img = crop_image(file)
        sum_img += 1
        # print(filename)
        cv2.imwrite('./cloud_images/original/'+filename+'.jpg',new_img)
print('done')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('./class_cloud/7/2021_11_16__12_00_05_HDR.jpg', cv2.COLOR_BGR2RGB)
r1 = image.shape[1]
r2 = image.shape[0]
mask = np.zeros((r2, r1), dtype=np.uint8)

cv2.circle(mask, (int(r1/2), int(r2/2)), 570, (255, 255, 255), -1, 8, 0)

masked = cv2.bitwise_and(image, image, mask=mask)
cropped = masked[120:1290, 450:1620]
plt.imshow(cropped)
cv2.imwrite('test_new_Crop.jpg',cropped)
```
